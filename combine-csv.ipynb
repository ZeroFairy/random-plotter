```python
import os
import pandas as pd
import numpy as np

# === CONFIGURATION ===
FOLDER_PATH = r"F:\Code\random-plotter\data\stationaryClustered\csv"
EXTENSION = ".csv"
OUTPUT_CSV = os.path.join("data", "stationaryClustered", "combine", "MEAN-combine_summary_report.csv")

# Ensure output directory exists
os.makedirs(os.path.dirname(OUTPUT_CSV), exist_ok=True)

# === STEP 1: Read all CSV files ===
all_files = [os.path.join(FOLDER_PATH, f) for f in os.listdir(FOLDER_PATH) if f.endswith(EXTENSION)]
dfs = [pd.read_csv(file) for file in all_files]

# === STEP 2: Combine all into a single DataFrame ===
combined_df = pd.concat(dfs, ignore_index=True)

# === STEP 3: Group and calculate mean ignoring zeros ===
numeric_cols = combined_df.select_dtypes(include="number").columns

def mean_ignore_zeros(series):
    """Return mean ignoring zeros, or NaN if all are zero."""
    non_zero = series[series != 0]
    return non_zero.mean() if not non_zero.empty else 0

agg_dict = {
    'Agent_movementModel': 'first',
    'Target_movementModel': 'first',
    **{col: mean_ignore_zeros for col in numeric_cols}
}

grouped = combined_df.groupby("filename").agg(agg_dict).reset_index()

# === STEP 4: Save result ===
grouped.to_csv(OUTPUT_CSV, index=False)
print("Saved to:", OUTPUT_CSV)
print(grouped)
```
